```python
import pandas as pd
import matplotlib.pyplot as plt 
import statsmodels.api as sm
```

```python
df = pd.read_csv('/workspaces/model_factory/groupe2/gr_2_test.csv', index_col=0)
df
```

```python
plt.figure(figsize=(12,5))
plt.title('Consommation d\'nergie renouvelable en France entre 1965 et 2021')
plt.plot(df['All_renew'], label='Consommation en TWh')
plt.legend(loc='best')
plt.show()
```

```python
df['Year'] = [i for i in range(1, len(df)+1)]
```

```python
model = sm.OLS.from_formula('All_renew ~ GDP_per_cap', data=df)
results = model.fit()
```

Tentons maintenant de comprendre le rsultat de notre modle. 

Les informations importantes sont: 

- Le **R_squared**: Le $R^2$ reprsente le pourcentage de variation expliqu par le modle. En termes simples, un $R^2 = 0$ correspond  un modle qui n'explique rien. Dans ce cas l, notre variable Year ne prdirerait strictement rien de notre consommation d'ER. Au contraire, un $R^2 = 1$ expliquerait tout. La totalit de la consommation d'ER se rsumerait  une fonction linaire du temps.
- Les valeurs de la colonne **coef**: Ici sont les valeurs qui nous intresse au dpart! La premire (intercept) correspond  $\alpha$. Cette valeur est le point de dpart de notre courbe. Pour l'anne 1965, la consommation d'ER tait donc de 47 Twh. La deuxime valeur (Year, correspond au $\beta$. La valeur tant de 0.8, cela implique que chaque anne, la consommation d'ER augmente de 0.8 Twh.
- Les valeurs de la colonne **P>|t|**: Ces valeurs nous indiquent de la significativit de nos coefficients. Simplement, pouvons nous croire que la valeur de notre coefficients $\alpha$ et $\beta$ sont rllement diffrent de 0. Une p-value valeur suffisament basse indique la significativit de notre coefficient. 

```python
results.summary()
```

Dans notre cas, nous avons un modle avec un beau $R^2$ de $0.563$ et des coefficients significatifs. Le $\beta$ est positif et donc indique la prsence d'une tendance linaire  la hausse.

# Un modle alternatif: La consommation d'nergie renouvelable suit une tendance quadratique en France

Un modle quadratique n'est pas bien plus compliqu d'un cours de mathmatiques en terminale. Nous remplaons simplement notre variable **Temps** par une variable **Temps^2**. 

Formellement: 
$$ All\_renew = \alpha + \beta * Temps^2  + \varepsilon $$ 

Il s'agit donc de rpter le processus prcdent en substituant **Temps** par **Temps^2**.

```python
df['Year_2'] = [i**2 for i in range(1,len(df)+1)]
```

```python
model_2 = sm.OLS.from_formula('All_renew ~ Year_2', data=df)
results_2 = model_2.fit()
```

Arretons-nous maintenant sur les rsultats de notre nouveau modle. 

1. Nous pouvons voir une augmentation du $R^2$  $62.9\%$. C'est une augmentation non-triviale. Ceci indique que notre modle correspond, d'une certaine manire, mieux  nos donnes.
2. La valeur des coefficients change ncessairement mais notre $\beta$ est toujours positif (ce qui indique une tendance de la consommation d'ER d'augmenter de plus en plus rapidement).
3. Nos coefficients sont toujours fortement significatifs. 

```python
results_2.summary()
```

**Comparons maintenant nos deux modles de manire visuelle**

```python
df['Prediction_2'] = results_2.predict()

plt.figure(figsize=(12,5))
plt.title('Prdiction de la consommation d\'nergie renouvelable en France entre 1965 et 2021')
plt.plot(df['All_renew'], label='Consommation en TWh')
plt.plot(df['Prediction_1'], label='Prediction linaire #1')
plt.plot(df['Prediction_2'], label='Prediction linaire #2')
plt.legend(loc='best')
plt.show()
```

Notre modle quadratique  l'air effectivement de nettement mieux reprsenter la tendance de consommation d'ER. Notre modle nous permet donc de dire que la consommation d'ER  l'air de s'acclrer depuis les annes 1970s. 

Il faut nanmoins rester vigilant dans notre interprtation. 

1. Nous sommes encores dans l'espace modle. Nous avons une prdiction mais notre modle ne nous apporte aucune explication quant aux raisons de cette tendance. A quoi est d ce comportement dans la consommation d'ER?

2. Regardons ce qu'il se passe si nous extrapolons les valeurs futures de la consommation d'ER: 

```python
n = 30 # number of predicting years 
future_years = pd.DataFrame(index=[i for i in range(2022,2022+n)], data={'Year_2': [i**2 for i in range(58,58+n)]})
predictions = results_2.predict(future_years)


plt.figure(figsize=(12,5))
plt.title('Prdiction de la consommation d\'nergie renouvelable en France entre 1965 et 2021')
plt.plot(df['All_renew'], label='Consommation en TWh')
plt.plot(df['Prediction_2'], label='Modle quadratique - in sample')
plt.plot(predictions, label='Modle quadratique - out sample')
plt.legend(loc='best')
plt.show()
```

Ce que l'on peut observer rapidemment, c'est que notre modle devient vite stupide. Il a identifi une tendance quadratique dans les donnes et ne prendra que cette dernire en compte pour prdire. Cela marchera peut-tre pour des prdictions  court terme, mais le modle deviendra obsolte pour les prdictions sur plusieurs dizaines d'annes.
